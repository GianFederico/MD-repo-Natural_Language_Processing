```python
# code for counting for each text:
# the total number of characters
# the total number of words
# the total number of sentences
# avg. chars per word
# avg. words per sentence
# avg. words repetitions in vocabulary (“lexical diversity”)
import nltk
from nltk.corpus import gutenberg
for fileid in gutenberg.fileids():
  num_chars = len(gutenberg.raw(fileid))
  num_words = len(gutenberg.words(fileid))
  num_sents = len(gutenberg.sents(fileid))

  avg_char=len(gutenberg.raw(fileid))/len(gutenberg.words(fileid))
  avg_word=len(gutenberg.words(fileid))/len(gutenberg.sents(fileid))
  all_voc_words= num_words/len(set([w.lower() for w in gutenberg.words(fileid)]))

  print ("FILE:",fileid, "->", num_chars, num_words, num_sents,"--",round(avg_char),round(avg_word),round(all_voc_words))
```

```python
import nltk
nltk.download('brown')
from nltk.corpus import brown

print(" categories->  ", brown.categories())
print(" fileids-> ", brown.fileids())
# print("raw:", brown.raw()) commented because exceedes jupyter data rate
print(" sents-> ", brown.sents())
print(" words-> ", brown.words())
print("_______________________________________________")

# A frequency distribution counts some occurrence, such as the use of a word or phrase.
news_text = brown.words(categories='news')
fdist = nltk.FreqDist(w.lower() for w in news_text)
modals = ['can', 'could', 'may', 'might', 'must', 'will']
for m in modals:
  print(m + ':', fdist[m], end=' ')

# A conditional frequency distribution, counts some occurrence separately for each of some number of conditions (Author, date, genre, etc.)
genre_word = [(genre, word) for genre in ['news', 'romance'] for word in brown.words(categories=genre)]
print ("\n_______________________________________________")
print (len(genre_word))
print (genre_word[0:10])

```

image.png

```python
#From the list of pairs we created through conditional freq dist, we can generate a conditional frequency distribution of words by genre

cfd = nltk.ConditionalFreqDist(genre_word)
print (cfd)
print (cfd.conditions())
print ("News:", cfd['news'])
print ("Romance: ",cfd['romance'])
print ("20 most common Romance:",
cfd['romance'].most_common(20))
print("Romance, could:", cfd['romance']['could'])
```

```python
import nltk
from nltk.corpus import brown
cfd = nltk.ConditionalFreqDist(
    (genre, word)
    for genre in brown.categories()
    for word in brown.words(categories=genre))
genres = ['news', 'religion', 'hobbies', 'science_fiction', 'romance', 'humor']
modals = ['can', 'could', 'may', 'might', 'must', 'will']
cfd.tabulate(conditions=genres, samples=modals)
```

```python
print ("Inaugural")
from nltk.corpus import inaugural #get the text
cfd = nltk.ConditionalFreqDist(
    (target, fileid[:4])  #the 2 axes
    for fileid in inaugural.fileids()
    for w in inaugural.words(fileid)  #all words in each file
    for target in ['america', 'citizen']
    if w.lower().startswith(target))  #narrow the choice
print (cfd['america'])
print (cfd['citizen'])
words = ['america', 'citizen']  #tabulating
years = [fileid[:4] for fileid in inaugural.fileids()]
cfd.tabulate(conditions=words, samples=years[:30])

print("_________________________")
print ("citizen, 1909: ",cfd['citizen']['1909']) #How many times was some version of “citizen” used in the 1909 inaugural address?
print ("america, 2009: ",cfd['america']['2009']) #How many times was “america” mentioned in 2009?
```

```python
#Define a function to compute what fraction of words in a text are not in the stopwords list

def content_fraction(text):
  stopwords= nltk.corpus.stopwords.words('english') #english stopwords list
  content=[w for w in text if w.lower() not in stopwords]
  return len(content)/len(text)

print ("Brown:", content_fraction(nltk.corpus.brown.words()), "of corpus total words are in the english stopword list")
```
