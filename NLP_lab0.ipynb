```python
#Define a function cfreq(wlist,txt) that computes the cumulative frequency of a list of words wlist in text txt

from nltk import *
def cfreq(word_list, text):
  sum=0
  for word in word_list:
    sum=sum + FreqDist(text).freq(word)
    print(word, FreqDist(text)[word])
  return sum

word_list=["whale","whaleman"]
cfreq(word_list, text1)
```

```python
#Define a function cfreq_n(txt,n) that computes the cumulative frequency of the n most common words in text txt

from nltk import *
def cfreq_n(text, n):
  fdist=FreqDist(text)
  combined_list=fdist.most_common(n)
  MC_list=[]
  freq=0
  for couple in combined_list:
    word=couple[0]
    count=couple[1]
    print(word,"->",count)

    MC_list.append(word)
    freq=freq + fdist.freq(word)

  return freq
  
cfreq_n(text1, 10)
```

```python
#Define a function cfreq_length(txt,k) that computes the cumulative frequency of the words in text txt having at least length k

def cfreq_length(text,k):
  fdist=FreqDist(text).keys()
  sum=0
  for word in fdist:
    if len(word) >= k:
      sum = sum + FreqDist(text).freq(word)
  return sum



cfreq_length(text1,10)
```

```python
#Define a function count_voc(txt,cf) that returns the number of most frequent items in the vocabulary that allows to achieve the cumulative frequency cf

def count_voc(text,cf):
  fdist = FreqDist(text)
  sortfreq = fdist.most_common()
  totfreq=0
  i=0
  while totfreq<cf:
    word= sortfreq[i][0]
    totfreq+=100*fdist.freq(word)
    i+=1
  return i

count_voc(text1,50)

#count_voc(text,50) returns how many tokens in the vocabulary are needed to achieve cumulative frequency of half of text 
```
